```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
apart_no = input("apartno >>>")
sample_data = pd.read_csv(apart_no+'.csv')
sample_data['date_time']= pd.to_datetime(sample_data.TS, unit='s')
sample_data['day'] = sample_data.date_time.dt.day
sample_data['month'] = sample_data.date_time.dt.month
sample_data['year'] = sample_data.date_time.dt.year
sample_data['time'] = sample_data.date_time.dt.time
sample_data['date'] =sample_data.date_time.dt.date
sample_data

temp = sample_data
temp['time'] = temp['time'].astype(str)
temp.loc[(temp.time=='00:00:00'), 'time'] = '00:30:00'
temp.loc[(temp.time=='01:00:00'), 'time'] = '01:30:00'
temp.loc[(temp.time=='02:00:00'), 'time'] = '02:30:00'
temp.loc[(temp.time=='03:00:00'), 'time'] = '03:30:00'
temp.loc[(temp.time=='04:00:00'), 'time'] = '04:30:00'
temp.loc[(temp.time=='05:00:00'), 'time'] = '05:30:00'
temp.loc[(temp.time=='06:00:00'), 'time'] = '06:30:00'
temp.loc[(temp.time=='07:00:00'), 'time'] = '07:30:00'
temp.loc[(temp.time=='08:00:00'), 'time'] = '08:30:00'
temp.loc[(temp.time=='09:00:00'), 'time'] = '09:30:00'
temp.loc[(temp.time=='10:00:00'), 'time'] = '10:30:00'
temp.loc[(temp.time=='11:00:00'), 'time'] = '11:30:00'
temp.loc[(temp.time=='12:00:00'), 'time'] = '12:30:00'
temp.loc[(temp.time=='13:00:00'), 'time'] = '13:30:00'
temp.loc[(temp.time=='14:00:00'), 'time'] = '14:30:00'
temp.loc[(temp.time=='15:00:00'), 'time'] = '15:30:00'
temp.loc[(temp.time=='16:00:00'), 'time'] = '16:30:00'
temp.loc[(temp.time=='17:00:00'), 'time'] = '17:30:00'
temp.loc[(temp.time=='18:00:00'), 'time'] = '18:30:00'
temp.loc[(temp.time=='19:00:00'), 'time'] = '19:30:00'
temp.loc[(temp.time=='20:00:00'), 'time'] = '20:30:00'
temp.loc[(temp.time=='21:00:00'), 'time'] = '21:30:00'
temp.loc[(temp.time=='22:00:00'), 'time'] = '22:30:00'
temp.loc[(temp.time=='23:00:00'), 'time'] = '23:30:00'
temp

for col in list(temp):
    if col in ['date_time']:
        temp=temp.drop(col,1)
temp

Data=temp
separator = ' '
Data['date_time'] = Data['date'].map(str) + separator + Data['time'].map(str)
Data['date_time']=pd.to_datetime(Data['date_time'])
Data

range = pd.date_range('2016-12-14 00:30:00', '2018-06-04 23:30:00', freq='1H')
df = pd.DataFrame(index = range)
df.reset_index(level=0, inplace=True)
df.rename(columns={'index':'date_time'},inplace=True)
df

result = pd.merge(df, Data, how='left')
result

data= result
data['day'] = data.date_time.dt.day
data['month'] =data.date_time.dt.month
data['year'] = data.date_time.dt.year
data['time'] = data.date_time.dt.time
data['power'] = data['W1']+ data['W2'] +data['W3']
data

def sum( yr, mnth ):
   
    yrwise = data[data['year'] == yr]
    monwise = yrwise[yrwise['month'] == mnth]
    df = monwise
    df = df.reset_index(drop=True)


    x=df['power']
    
    #z=0.3 + 0.7*(x-x.min())/(x.max()-x.min())
    #df['norm']=z
   
    d = monwise.day.unique()
    days = d.tolist()
    totdays = len(days)

    mylist = df['power'].tolist()

    myarray = np.asarray(mylist)

    myarray= myarray.reshape(totdays,24)
    myarray.shape

    mydf=pd.DataFrame(myarray)
    mydf

    mydf = mydf.interpolate(method='index', axis=0).ffill(axis=0).bfill(axis=0)
    mydf = mydf.interpolate(method='index', axis=1).ffill(axis=1).bfill(axis=1)
    
    bbb=mydf.describe()
    x=bbb.loc['mean',:]
    mydf.loc[totdays]=x

    y=mydf.transpose()

    y.rename(columns={totdays:'avg'},inplace=True)
            
    bcd=y.describe()
    sss=bcd.loc['mean',:]
    #sss=sss.round(2)
    y.loc['d_avg']=sss
    y.at['d_avg','avg'] = 'nan'
    #k = y.round(2)
    return y;

with pd.ExcelWriter('Apt'+apart_no+'_nnp.xlsx') as writer:
    m1= sum(2016,12)
    m1.to_excel(writer, sheet_name='Dec16')
    m2= sum(2017,1)
    m2.to_excel(writer, sheet_name='Jan17')
    m3= sum(2017,2)
    m3.to_excel(writer, sheet_name='Feb17')
    m4= sum(2017,3)
    m4.to_excel(writer, sheet_name='Mar17')
    m5= sum(2017,4)
    m5.to_excel(writer, sheet_name='Apr17')
    m6= sum(2017,5)
    m6.to_excel(writer, sheet_name='May17')
    m7= sum(2017,6)
    m7.to_excel(writer, sheet_name='Jun17')
    m8= sum(2017,7)
    m8.to_excel(writer, sheet_name='Jul17')
    m9= sum(2017,8)
    m9.to_excel(writer, sheet_name='Aug17')
    m10= sum(2017,9)
    m10.to_excel(writer, sheet_name='Sep17')
    m11= sum(2017,10)
    m11.to_excel(writer, sheet_name='Oct17')
    m12= sum(2017,11)
    m12.to_excel(writer, sheet_name='Nov17')
    m13= sum(2017,12)
    m13.to_excel(writer, sheet_name='Dec17')
    m14= sum(2018,5)
    m14.to_excel(writer, sheet_name='May18')
    m15= sum(2018,6)
    m15.to_excel(writer, sheet_name='Jun18')
```
